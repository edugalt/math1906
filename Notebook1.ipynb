## 1.2 Newtonian description of motion (and Science!) 

<br><br><br>

1.How to compute motion:

$$F = ma = m \ddot{x} \equiv m \frac{d^2 x}{dt^2}$$

2.Expression for the force $F$:
$$ F_G = -G \frac{mM}{r^2}$$

3.Result: unify earth and sky!

### New interpretation of motion (Science):

<font color=red> The motion (curve) is the solution of an equation of motion (a rule how to move):

$$ m \ddot{x} = F(x) $$

This is a <i> differential equation, </i> a form of dynamical system.</font>

<br><br><br><br><br><br>

## Assumed implications for Science (Physics)

The behaviour of matter results from the forces acting on it. 

1) Knowing the forces (equation of motion) we know everything about the future evolution of the system.

2) The goal of Physics is to reveal the fundamental forces of nature.
